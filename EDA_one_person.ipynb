```python
import pandas as pd
data = pd.read_csv('person.csv',index_col=0)
data.head()
```

```python
import numpy as np
resp_corr = pd.DataFrame(data.drop(['Time [s]','sec'],axis=1).corr()[' RESP'])
resp_corr.reset_index(inplace=True)
resp_corr['abs'] = resp_corr[' RESP'].apply(lambda x: np.abs(x))
resp_corr['resp'] = resp_corr['index'].apply(lambda x: 1 if 'RESP' in x else 0)
resp_corr = resp_corr[resp_corr['resp']==0]
resp_corr.sort_values(by='abs',ascending=False,inplace=True)
resp_corr.drop(['resp','abs'],axis=1,inplace=True)
resp_corr
```
